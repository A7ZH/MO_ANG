```
import pandas as pd
from ast import literal_eval as make_tuple
import folium
from IPython.display import display
```

```
!rm -rdf sample_data
# Getting the Data
!git clone https://github.com/A7ZH/Restaurant_Intelligence.git &> /dev/null
df = pd.read_csv("./Restaurant_Intelligence/SkipTheDishes/SkipTheDishes_Clean.csv", 
                 encoding='utf-8-sig')
df
```

```
cuisine_map = {}
for ind, row in df.iterrows():
  cuisines = row['Cuisine'].split(';')
  coordinate = make_tuple(row['Coordinate'])
  for c in cuisines:
    c = c.lstrip().rstrip()
    if c in cuisine_map:
      cuisine_map[c] += [coordinate] 
    else:
      cuisine_map[c] = [coordinate]
```

```
def plot_cuisine_map(*cuisine_names):
                             # Center of Downtown Toronto
  map = folium.Map(location=[43.653963, -79.387207],
                   tiles='OpenStreetMap',#'cartodbpositron',
                   zoom_start=12)
  for cuisine_name in cuisine_names:
    for coordinate in cuisine_map[cuisine_name]:
      folium.CircleMarker(location=coordinate).add_to(map)
  display(map)
```

```
# Type in any number of cuisines separated by comma
plot_cuisine_map("Indian", "Indien") 
```
